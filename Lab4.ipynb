```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import datasets
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.model_selection import KFold, cross_val_score
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
x = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]
```

```python
X_scaled = scale(x)
```

```python
kf = KFold(n_splits=5, random_state=42, shuffle=True)
metrics = np.linspace(1, 10, num=200)
means = {}
for p in metrics:
    knn = KNeighborsRegressor(p=p, n_neighbors=5, weights='distance')
    score = cross_val_score(knn, X_scaled, y, cv=kf,
                            scoring='neg_mean_squared_error')
    score_mean = np.mean(score)
    means[p] = score_mean
df = pd.DataFrame(list(means.items()), columns=['p', 'score_mean'])
df.head(10)
```

```python
df[df['score_mean'] == df['score_mean'].max()]
```
