```python
import pandas as pd
import numpy as np
from scipy. stats import pearsonr
from sklearn.tree import DecisionTreeClassifier


data = pd.read_csv('/content/drive/MyDrive/titanic.csv',
                   index_col='PassengerId')
```

```python
data.head(10)
```

```python
df = data[['Pclass', 'Fare', 'Age', 'Sex', 'Survived']].dropna()
df['Sex'].loc[df['Sex'] == 'female'] = 1
df['Sex'].loc[df['Sex'] == 'male'] = 0
df
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(df[['Pclass', 'Fare', 'Age', 'Sex']], df[['Survived']])
```

```python
important = pd.Series(clf.feature_importances_,
                      index=list(df[['Pclass', 'Fare', 'Age', 'Sex']]))
print(' '.join(important.sort_values(ascending=False).head(2).index.values))
```
